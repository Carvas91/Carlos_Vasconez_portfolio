```python
# Supongamos que tienes un DataFrame 'df' y deseas crear una nueva columna 'Mes_Num' con los valores numéricos de los meses.

meses = {
    'enero': '01',
    'febrero': '02',
    'marzo': '03',
    'abril': '04',
    'mayo': '05',
    'junio': '06',
    'julio': '07',
    'agosto': '08',
    'septiembre': '09',
    'octubre': '10',
    'noviembre': '11',
    'diciembre': '12'
}

# Aplicamos el mapeo a la columna 'Mes_i' y creamos la nueva columna 'Mes_Num'
df['Mes_Num'] = df['Mes_i'].str.lower().map(meses)

# Convierte 'Mes_Num' a tipo de dato numérico (entero)
df['Mes_Num'] = df['Mes_Num'].astype(int)

# Ahora, 'df' contiene una nueva columna 'Mes_Num' con los valores numéricos de los meses.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import calendar

# Supongamos que tienes un DataFrame 'df' con las columnas 'Mes_Num' y 'Tipo Muert.'
# y 'Mes_Num' es una columna numérica que representa los meses.

# Creamos el gráfico de serie temporal con estilo mejorado
plt.figure(figsize=(12, 6))
ax = df.groupby('Mes_Num')['Tipo Muert.'].count().plot(kind='line', marker='o', color='b', linestyle='-', linewidth=2)

# Personalizamos el gráfico
plt.title('Serie Temporal de Tipos de Muerte por Mes', fontsize=16)
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Cantidad', fontsize=12)
plt.xticks(range(1, 13), calendar.month_abbr[1:], rotation=45)
plt.grid(True)

# Agregamos etiquetas a los puntos del gráfico
for x, y in zip(df['Mes_Num'], df.groupby('Mes_Num')['Tipo Muert.'].count()):
    plt.text(x, y, str(y), ha='right', va='bottom', fontsize=10)

# Mostramos la leyenda
plt.legend(['Cantidad de Muertes'], loc='upper right')

# Mostramos el gráfico
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Asumiendo que 'df' es tu DataFrame y que ya está cargado con los datos

# Convertir las fechas a formato datetime de Python
df['Fecha Infracción'] = pd.to_datetime(df['Fecha Infracción'], format='%d/%m/%Y')

# Ordenar el DataFrame por la fecha
df.sort_values('Fecha Infracción', inplace=True)

# Mapear cada sexo a un color
# Continuando con el código...

# Mapear cada sexo a un color
colors = {'HOMBRE': 'blue', 'MUJER': 'red', 'OTRO': 'green'}  # Ajusta según corresponda a tus datos

# Crear el gráfico de dispersión
plt.figure(figsize=(10, 6))

# Dibujar cada punto en el gráfico
for sexo in df['Sexo'].unique():
    subset = df[df['Sexo'] == sexo]
    plt.scatter(subset['Fecha Infracción'], subset['Edad'],
                color=colors.get(sexo, 'black'), # Color por defecto en caso de no encontrar el sexo
                label=sexo)

# Formatear el eje x para mostrar las fechas adecuadamente
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d/%m/%Y'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())

# Rotar las fechas para que se vean mejor
plt.gcf().autofmt_xdate()

plt.xlabel('Fecha de Infracción')
plt.ylabel('Edad')
plt.title('Gráfico de dispersión de Edades por Sexo y Fecha de Infracción')
plt.legend()
plt.show()
```

```python
#distribucion de sexos de las victimas
df.Sexo.value_counts()
```

```python
#calculo de la probabilidad de que la victima sea un hombre P(Hombre)
probabilidad_sexo = df.Sexo.value_counts() / len(df)
probabilidad_hombre = probabilidad_sexo[0]
probabilidad_hombre
#una victima tiene 92% probabilidad de ser hombre
```

```python
#calculo probabilidad de que la victima sea de X etnia y hombre P(Etnia & Hombre)
filtered_df = df[df['Sexo'] == 'HOMBRE']
ethnicity_counts = filtered_df['Etnia'].value_counts().sort_values(ascending=False)
prob_etnia_hombre = ethnicity_counts / len(df)
prob_etnia_hombre
```

```python
#probabilidad condicional de que la victima sea de X etnia  dado de que es hombre
prob_cond_etnia_hombre =prob_etnia_hombre/probabilidad_hombre
prob_cond_etnia_hombre
```
